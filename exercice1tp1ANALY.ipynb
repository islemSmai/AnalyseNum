```python
import matplotlib.pyplot as plt
from numpy import linspace
f=lambda x: (x-1)**6
T=linspace(-2,2,41)
plt.title("solution dans [-2:2] ");
plt.xlabel("x")
plt.ylabel ("f(x)")
plt.grid(True)
plt.plot(T,f(T))
```

```python
f=lambda x: x**3+x**2-3*x-3
def dicho (f , a , b , epsilon ):
    g , d = min (a , b ) , max (a , b )
    while d - g > epsilon :
        m = ( g + d ) / 2
        if f ( m ) == 0:
            return m
        elif f ( g ) * f ( m ) < 0:
            d = m
        else :
            g = m
        return ( g + d ) / 2

    
        

    
```
